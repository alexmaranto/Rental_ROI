```python
class RentalProperty:
    def __init__(self, rental_income, taxes, insurance, vacancy, repairs, capex, mortgage, prop_mgmt):
        self.rental_income = rental_income
        self.taxes = taxes
        self.insurance = insurance
        self.vacancy = vacancy
        self.repairs = repairs
        self.capex = capex
        self.mortgage = mortgage
        self.prop_mgmt = prop_mgmt

    def calculate_expenses(self):
        return self.taxes + self.insurance + self.vacancy + self.repairs + self.capex + self.mortgage + self.prop_mgmt

    def calculate_cash_flow(self):
        return self.rental_income - self.calculate_expenses()

class ReturnOnInvestment:
    def __init__(self, down_payment, closing_cost, rehab_budget):
        self.down_payment = down_payment
        self.closing_cost = closing_cost
        self.rehab_budget = rehab_budget

    def calculate_total_investment(self):
        return self.down_payment + self.closing_cost + self.rehab_budget

    def calculate_cash_on_cash_roi(self, annual_cash_flow):
        return (annual_cash_flow / self.calculate_total_investment()) * 100

if __name__ == "__main__":
    rental_property = RentalProperty(
        rental_income= float(input("Enter the monthly rental income: $")),
        taxes= float(input("Enter the monthly taxes due: $")),
        insurance= float(input("Enter the insurance payment: $")),
        vacancy= float(input("Enter the vacancy expense: $")),
        repairs= float(input("Enter the monthly repair expense: $")),
        capex= float(input("Enter the CapEx expense: $")),
        mortgage= float(input("Enter the monthly mortgage payment: $")),
        prop_mgmt= float(input("Enter the monthly property management expense: $"))
    )

    monthly_expenses = rental_property.calculate_expenses()
    monthly_cash_flow = rental_property.calculate_cash_flow()

    print(f"\nCategory: Income")
    print(f"Rental Income: ${rental_property.rental_income:.2f}")

    print("\nCategory: Expenses")
    print(f"Taxes: ${rental_property.taxes:.2f}")
    print(f"Insurance: ${rental_property.insurance:.2f}")
    print(f"Vacancy: ${rental_property.vacancy:.2f}")
    print(f"Repairs: ${rental_property.repairs:.2f}")
    print(f"CapEx: ${rental_property.capex:.2f}")
    print(f"Mortgage: ${rental_property.mortgage:.2f}")
    print(f"Property Management: ${rental_property.prop_mgmt:.2f}")
    print(f"Total Monthly Expenses: ${monthly_expenses:.2f}")

    print("\nCategory: Cash Flow")
    print(f"Income - Expenses = Cash Flow")
    print(f"Cash Flow: ${monthly_cash_flow:.2f}")

    roi = ReturnOnInvestment(
        down_payment= float(input("Enter the down payment of the property: $")),
        closing_cost= float(input("Enter the closing cost: $")),
        rehab_budget= float(input("Enter the rehab budget: $"))
    )

    annual_cash_flow = monthly_cash_flow * 12
    cash_on_cash_roi = roi.calculate_cash_on_cash_roi(annual_cash_flow)

    print("\nCash on Cash ROI")
    print(f"Down Payment: ${roi.down_payment}")
    print(f"Closing Cost: ${roi.closing_cost:.2f}")
    print(f"Rehab Budget: ${roi.rehab_budget:.2f}")
    print(f"Total Investment: ${roi.calculate_total_investment():.2f}")
    print(f"Annual Cash Flow: ${annual_cash_flow:.2f}")
    print(f"Annual Cash Flow / Total Investment = {cash_on_cash_roi:.1f}%")
```
